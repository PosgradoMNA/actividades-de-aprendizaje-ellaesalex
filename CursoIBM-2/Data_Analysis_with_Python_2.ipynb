### **Data Normalization in Python**

- Técnica para comprender en el preprocesamiento de datos
- Al hacer que los rangos sean consistentes entre las variables, la normalización permite una comparación más justa entre las diferentes características, al asegurarse que tengan el mismo impacto
- Técnicas de normalización
  - Escalado de características simple (Simple Feature Scaling)
    - Se divide cada valor por el
valor máximo para esa función
    - Esto hace que los nuevos valores oscilen entre 0 y 1
  - Min-Max
    - Toma cada valor, X_old, restándolo del mínimo valor de esa característica, luego se divide por el rango de esa característica    
    - Esto hace que los nuevos valores oscilen entre 0 y 1
  - Puntaje z o Puntaje estándar
    - Para cada valor, resta el Mu que es el promedio de la característica,y luego se divide por la desviación estándar (sigma)
    - Los valores resultantes oscilan alrededor de 0 y, por lo general, oscilan entre -3 y +3, pero pueden ser más altos o bajo

**En Python**
Se puede aplicar el método de normalización en la "longitud"
rasgo
- Escalado de características simple (Simple Feature Scaling)
  - Usando el método Max

```
df["column"] = df["column"]/df["column"]. max()
```

- Min-Max

```
df["column"] = (df["column"]-df["column"]. min())/
               (df["column"]-df["column"]. max()-df["column"]-df["column"]. min())
```

- Puntaje z o Puntaje estándar

```
df["column"] = (df["column"]-df["column"]. mean()) / 
               df["column"].std()
```

### **Binning in Python**

- Método de prepricesamiento de datos
- Es cuando se agrupam valores en contenedores
  - Ej: Edad en [0 a 5],[6 a 10],[11, 15]
- Puede mejorar la precisión de los modelos predictivos
- Se puede agrupar un conjunto de valores numéricos en un menor número de contenedores para tener una mejor comprensión de la distribución de datos
  - Ej: Rango de 4500 a 5000 en las categorías "Precio Bajo", "Precio Medio", "Precio Alto"


**En Python**
```
bins = np.linspace(min(df["column"]),max(df["column"]),4)

group_names = ["Low", "Medium", "High"]

df["new-column"] = pd.cut(df["column"],bins, labels = group_names,include_lowest = True)
```
