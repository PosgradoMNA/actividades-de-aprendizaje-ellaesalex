**groupby() method**
- Se encuentra en Pandas
- Se usa en variables categóricas
- Agrupa los datos en subconjuntos según a las diferentes categorías de esa variable
- Puede agrupar por una sola variable o puede agrupar por múltiples variables pasando
en varios nombres de variables

Selección de las columnas a interés
```
df_test = df [['columna1, 'columna2', 'columna3']]
```

Agrupación de los datos reducidos según 'columna1' y 'columna2'. Se agrega la media de la 'columna3' al final
```
df_grp = df_test.groupby (['columna1, 'columna2'], as_index=False).mean()
df_grp
```


**Pivot method**

- En Pandas
- Tabla dinámica 
- Tiene una variable mostrada a lo largo de las columnas y la otra variable mostrada a lo largo de las filas
```
df_pivot = df_grp.pivot(index = 'columna1', columns = 'bodu-style')
```



**Heat map**

**Métodos estadísticos de correlación**

- Correlación de Pearson
  - Mide la fuerza de la correlación entre variables numéricas continuas
  - Da:
    - Coeficiente de correlación
      - Valor cercano a 1: Gran correlación positiva
      - Valor cercano a -1: Gran correlación negativa 
      - Valor cercano a 0: Implica que no hay correlación entre las variables
    - P-value
      - Dirá cuán seguros estamos acerca de la correlación que calculamos
      - Valor inferior a 0,001: Gran certeza sobre la correlación coeficiente que calculamos
      - Valor entre 0,001 y 0,05: Certeza moderada
      - Valor entre 0,05 y 0,1: Certeza débil
      - Valor mayor que 0.1: Nada de certeza

- Correlación fuerte
   - Coeficiente de correlación es cercano a 1 o -1 
   - P-value es menor que 0.001


**En Python con Scipy**
```
pearson_coef, p_value = stats.pearsonr(df['columna1'], df['columna2'])
```
