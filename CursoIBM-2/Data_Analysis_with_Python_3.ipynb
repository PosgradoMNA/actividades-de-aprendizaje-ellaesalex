### **Exploratory Data Analysis**

- EDA

### **Descriptive Statistics**

- Describe las características básicas de un conjunto de datos y obtiene un breve resumen sobre la muestra y las medidas de los datos

**Métodos útiles**
- describe() en Pandas
  - Calcula automáticamente las estadísticas básicas para todas las variables numéricas
  - Muestra la media, el número total de datos, la desviación estándar, los cuartiles y los valores extremos


**Box Plots**
- Diagrama de caja
- Ayudan a  visualizar datos con valores numéricos
- Se visualizar las distintas distribuciones de los datos
- Muestra la mediana de los datos, que representa dónde es el punto de datos medio
- Cuartil superior: Muestra dónde se encuentra el percentil 75
- Cuartil inferior muestra dónde está el percentil 25
- Los datos entre la parte superior y el cuartil inferior representa el rango intercuartílico
- Valores atípicos: Son los puntos individuales que ocurren fuera de la parte superior y inferiores de los extremos

  - Variable predictora (Predictor/independent variable)

    - Se está utilizando para predecir un resultado
    - Ej: Tamaño del motor de un carro
  - Variable objetivo (Target/dependent variable)
    - On y-axis
    - Está tratando de predecir.
    - Ej: Precio de un carro


 

### **GroupBy in Python**

**groupby() method**
- Se encuentra en Pandas
- Se usa en variables categóricas
- Agrupa los datos en subconjuntos según a las diferentes categorías de esa variable
- Puede agrupar por una sola variable o puede agrupar por múltiples variables pasando
en varios nombres de variables

Selección de las columnas a interés
```
df_test = df [['columna1, 'columna2', 'columna3']]
```

Agrupación de los datos reducidos según 'columna1' y 'columna2'. Se agrega la media de la 'columna3' al final
```
df_grp = df_test.groupby (['columna1, 'columna2'], as_index=False).mean()
df_grp
```


**Pivot method**

- En Pandas
- Tabla dinámica 
- Tiene una variable mostrada a lo largo de las columnas y la otra variable mostrada a lo largo de las filas
```
df_pivot = df_grp.pivot(index = 'columna1', columns = 'bodu-style')
```



**Heat map**

- Mapa de calor
- Toma una cuadrícula rectangular de datos y asigna una intensidad de color basada en el valor de los datos en los puntos de la cuadrícula
- Traza la variable de destino en múltiples variables y, a través de esto, obtener pistas visuales de la relación entre estas variables y el objetivo

**pcolor de Pylot**
```
plt.pcolor(df_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```
