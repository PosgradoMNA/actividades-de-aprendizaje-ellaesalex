### **Model Development**

- Modelo o estimador como una ecuación matemática utilizada para predecir un valor dado uno o más valores
- Relacionar una o más variables o características independientes con variables dependientes
- Cuantos más datos relevantes tenga, más preciso será su modelo.

### **Linear Regression and Multiple Linear Regression**

**Regresión lineal Simple**
- SLR
- Se refiere a una variable independiente para hacer una predicción
- Método que ayuda a entender la relación entre dos variables: la variable predictora (independiente) x y la variable objetivo (dependiente) y
- y = b0 + b1 x
- bo = Intercepto
- b1 = Pendiente
- Almacenamos los puntos de datos en dos marcos de datos o matrices numpy
- El valor que nos gustaría predecir se llama el **objetivo** que almacenamos en la matriz y, almacenamos la variable dependiente en el marco de datos o matriz X
- Si la suposición lineal es correcta, este error se debe al ruido

Proceso
- Tenemos un conjunto de puntos de entrenamiento 
- Usamos estos puntos de entrenamiento para ajustar o entrena el modelo y obtener parámetros 
- Luego usamos estos parámetros en el modelo
- Ahora tenemos un modelo; usamos el sombrero para indicar que el modelo es una estimación
- Podemos usar este modelo para predecir valores que no hemos visto

**Regresión lineal múltiple**
- Se refiere a múltiples variables independientes para hacer una predicción
- Se utiliza para explicar la relación entre una target (Y) variable, y Two or more predictor (X) variables
- Ejemplo
  - Si tenemos por ejemplo 4 variables predictoras, entonces:
    - B0: intercepto (X=0) 
    - B1: el coeficiente o parámetro de 𝑋1
    - B2: el coeficiente del parámetro 𝑋2
- Las variables 𝑋1 y 𝑋2 se pueden visualizar en un plano 2D

Proceso

```
Z = df[['horsepower','curb-weight', 'engine-size', 'highway-mog']]
```

```
lm.fit (Z, df['price'])
```

```
Yhat = lm.predict(X)
```

**Elaboración de un gráfico de distribución (Distribution Plot)**
- Cuenta el valor predicho frente al valor real
- Útiles para visualizar modelos con más de una variable independiente
o característica
- Los valores de los objetivos y los valores predichos son continuos
- Un histograma es para valores discretos
- Pandas los convertirá en una distribución
- El acceso vertical se escala para que el área bajo la distribución sea igual a uno
- Distribución: hist = False
- Histograma: hist = True

```
import = seaborn as sns

axl = sns.distplot(df['price'], hist = False, color = "r", label = "Actual Value")
sns.distplot (Yhat, hist = False, color = "b", label = "Fitted Values"
ax=axl)
```

### **Polynomial Regression and Pipelines**

**Regresión polinomial**
- Método es beneficioso para describir relaciones curvilíneas
- Pipelines: Son una forma de simplificar el código
- Relación curvilínea: Es el resultado elevando al cuadrado o estableciendo términos de orden superior de las variables predictoras en el modelo, transformando los datos
- Modelo cuadrático: La variable predictora del modelo está elevada al cuadrado
- Regresión polinomial de tercer orden: El modelo puede ser cúbico, lo que significa que la variable predictora está al cubo
- Regresiones polinómicas de orden superior
- Al cambiar el orden del polinomio, el grado de regresión hace una gran diferencia y puede resultar en un mejor ajuste si elige el valor correcto
- La relación entre la variable y el parámetro es siempre lineal
